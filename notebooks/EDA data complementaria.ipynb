```python
# Seleccionar las columnas para el análisis univariado
columnas = ['AIDS Report Number', 'Local Event Date', 'Event City', 'Event State',
       'Event Airport', 'Event Type', 'Aircraft Damage', 'Flight Phase',
       'Aircraft Make', 'Aircraft Model', 'Aircraft Series', 'Operator',
       'Primary Flight Type', 'Flight Conduct Code', 'Flight Plan Filed Code',
       'Aircraft Registration Nbr', 'Total Fatalities', 'Total Injuries',
       'Aircraft Engine Make', 'Aircraft Engine Model', 'Engine Group Code',
       'Nbr of Engines', 'PIC Certificate Type', 'PIC Flight Time Total Hrs',
       'PIC Flight Time Total Make-Model', ' ', ' .1']

# Obtener el resumen estadístico de las columnas numéricas
resumen_estadistico = data[columnas].describe()

# Calcular la moda de las columnas categóricas
moda = data[columnas].mode()

# Calcular la tabla de frecuencia para las columnas categóricas
tabla_frecuencia = pd.DataFrame()
for col in columnas:
    if data[col].dtype == 'object':  # Considerar solo las columnas categóricas
        tabla_frecuencia[col] = data[col].value_counts()

# Mostrar los resultados
print("Resumen estadístico:")
print(resumen_estadistico)
print("\nModa:")
print(moda)
print("\nTabla de frecuencia:")
print(tabla_frecuencia)
```

```python
# Seleccionar las columnas numéricas de interés
columnas_numericas = ['Total Fatalities', 'Total Injuries','Nbr of Engines']

# Crear un nuevo DataFrame con las columnas numéricas
data_numericas = data[columnas_numericas]

# Convertir las columnas a tipo numérico
data_numericas = data_numericas.astype(float)

# Calcular la matriz de correlación para las columnas numéricas
correlation_matrix = data_numericas.corr()

# Mostrar la matriz de correlación
print("Matriz de correlación:")
print(correlation_matrix)
```
