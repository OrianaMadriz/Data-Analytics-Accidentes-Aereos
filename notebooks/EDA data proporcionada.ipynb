```python
#Importar las librerias necesarias
import pandas as pd
import numpy as np
```

```python
# Cargar el dataset
data=pd.read_csv("AccidentesAviones.csv")
```

```python
# Explorar el dataset
# Visualizar las primeras filas del DataFrame
data.head(3)
```

```python
# Obtener informaci√≥n del dataset
print(data.info())
```

```python
# Verificar duplicados
data.duplicated().value_counts()
```

```python
# Seleccionar las columnas para el an√°lisis univariado
columnas = ["ID",'fecha', 'HORA_declara', 'Ruta', 'OperadOR', 'flight_no', 'route',
            'ac_type', 'registration', 'cn_ln', 'all_aboard', 'PASAJEROS_A_BORDO',
            'crew_aboard', 'cantidad_de_fallecidos', 'passenger_fatalities',
            'crew_fatalities', 'ground', 'summary']

# Obtener el resumen estad√≠stico de las columnas num√©ricas
resumen_estadistico = data[columnas].describe()

# Calcular la moda de las columnas categ√≥ricas
moda = data[columnas].mode()

# Calcular la tabla de frecuencia para las columnas categ√≥ricas
tabla_frecuencia = pd.DataFrame()
for col in columnas:
    if data[col].dtype == 'object':  # Considerar solo las columnas categ√≥ricas
        tabla_frecuencia[col] = data[col].value_counts()

# Mostrar los resultados
print("Resumen estad√≠stico:")
print(resumen_estadistico)
print("\nModa:")
print(moda)
print("\nTabla de frecuencia:")
print(tabla_frecuencia)
```

```python
#En el analisis exploratorio no se deberia modificar el dataset, solo observarlo, pero para fines de exploracion de los datos se reemplaza los "?" por valores nulos y asi poder aplicar diferentes herramientas de exploraci√≥n.
columnas=['ID', 'fecha', 'HORA_declara', 'Ruta', 'OperadOR', 'flight_no', 'route',
       'ac_type', 'registration', 'cn_ln', 'all_aboard', 'PASAJEROS_A_BORDO',
       'crew_aboard', 'cantidad_de_fallecidos', 'passenger_fatalities',
       'crew_fatalities', 'ground', 'summary']
# Reemplazar los valores no num√©ricos por NaN
data[columnas] = data[columnas].replace('?', np.nan)
```

```python
# Seleccionar las columnas num√©ricas de inter√©s
columnas_numericas = ['all_aboard', 'PASAJEROS_A_BORDO', 'crew_aboard', 'cantidad_de_fallecidos', "passenger_fatalities","crew_fatalities",  "ground"]

# Crear un nuevo DataFrame con las columnas num√©ricas
data_numericas = data[columnas_numericas]

# Convertir las columnas a tipo num√©rico
data_numericas = data_numericas.astype(float)

# Calcular la matriz de correlaci√≥n para las columnas num√©ricas
correlation_matrix = data_numericas.corr()

# Mostrar la matriz de correlaci√≥n
print("Matriz de correlaci√≥n:")
print(correlation_matrix)
```
